```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
digits = load_digits()
X, y = digits.data, digits.target
```

```python
print("Dataset shape:", X.shape)
print("Target shape:", y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(X_train, y_train)
y_pred1 = log_reg.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, ConfusionMatrixDisplay
```

```python
acc1 = accuracy_score(y_test, y_pred1)
print("Logistic Regression Accuracy:", acc1)
```

```python
ConfusionMatrixDisplay.from_estimator(log_reg, X_test, y_test, cmap="Blues")
plt.title("Logistic Regression Confusion Matrix")
plt.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
y_pred2 = rf.predict(X_test)
```

```python
acc2 = accuracy_score(y_test, y_pred2)
print("Random Forest Accuracy:", acc2)
```

```python
ConfusionMatrixDisplay.from_estimator(rf, X_test, y_test, cmap="Greens")
plt.title("Random Forest Confusion Matrix")
plt.show()
```
